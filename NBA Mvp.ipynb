```python
#read in data
df = pd.read_csv('nbamvp2020.csv')

#clean up player names
#df['Player (Team)'] = df['Player (Team)'].str.rsplit(" ",expand=True)[::-1]

#transpose the data - This is done so we can plot everything correctly
df = df.T

#Change column names to be the player names
df.columns = df.iloc[0]

#We want to drop the player names row as well as the total points row
df = df.drop('Player (Team)')
df = df.drop('Total Points')

#Create a new list of columns to recreate the dataframe to be exactly how we want it
my_list = df.columns.values.tolist()
df = pd.DataFrame(df,columns=my_list)
df
```

```python
fig = plt.figure(
    FigureClass = Waffle,
    rows = 10,
    labels = list(df.index),
    values = df.iloc[:,0],
    legend = {
        'loc': 'lower left'
    },
    title={
        'label': 'Giannis',
    }
)
```
